```
from ecies.utils import generate_eth_key, generate_key
from ecies import encrypt, decrypt
```

```
eth_k = generate_eth_key()
# hex string 
sk_hex = eth_k.to_hex()

print(sk_hex)
# hex string
pk_hex = eth_k.public_key.to_hex()  
print(pk_hex)
```

```
data = b'Emma graduation on 07/05/2020 7pm at my house'
```

```
encripted_msg = encrypt(pk_hex, data)
print(encripted_msg)
```

```
decipted_msg = decrypt(sk_hex, encripted_msg)
print(decipted_msg)
```
